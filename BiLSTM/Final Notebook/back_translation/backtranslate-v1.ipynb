```python
import time
# zh_hate_df=dataset[(dataset.language == 3) & (dataset.label == 1)]
id_hate_df=dataset[(dataset.language == 13) & (dataset.label == 1)]


# Define your function to just return values, not print them
def back_translate(text, translator1, translator2):
    try:
        translated1 = translator1.translate(text)
        translated2 = translator2.translate(translated1)
        print(translated2, "==", text)
        time.sleep(3)
        return translated1, translated2
    except:
        return "ERROR", "ERROR"

# Function to process DataFrame and append to CSV
def process_and_cache(df, translator1, translator2, filename):
    columns = df.columns.tolist() + ['translated_text', 'back_translated_text']
    for index, row in df.iterrows():
        translated_text, back_translated_text = back_translate(row['text'], translator1, translator2)
        new_row = pd.DataFrame([row.tolist()+[translated_text, back_translated_text]], columns=columns)
        # Append to CSV file, creating it if it does not exist, else appending
        new_row.to_csv(filename, mode='a', header=not pd.io.common.file_exists(filename), index=False)

# Assuming zh_en_translator and en_zh_translator are defined
process_and_cache(zh_hate_df, zh_en_translator, en_zh_translator, 'kaggle/input/multilingual-hatespeech-dataset/Dataset/Training/zh_hate_translations.csv')
process_and_cache(id_hate_df, id_en_translator, en_id_translator, 'kaggle/input/multilingual-hatespeech-dataset/Dataset/Training/id_hate_translations.csv')

    
    
```
